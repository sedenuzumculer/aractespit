```python
import cv2
import json
import numpy as np
import matplotlib
matplotlib.use('TkAgg') 
import matplotlib.pyplot as plt





# Tıklama olaylarını işleyen fonksiyon
def mouse_callback(event, x, y, flags, param):
    global parking_spaces, current_space

    if event == cv2.EVENT_LBUTTONDOWN:
        current_space.append((x, y))
        cv2.circle(img, (x, y), 5, (0, 0, 255), -1)
        if len(current_space) == 4:
            parking_spaces.append(current_space)
            cv2.fillPoly(img, [np.array(current_space, np.int32)], (0, 255, 0))
            current_space = []

# Görüntüyü yükleyin
img = cv2.imread('/Users/sedenuzumculer/aractespit/IMG_5498.jpg')

# Matplotlib figürü oluşturun
fig = plt.figure()
ax = fig.add_subplot(111)
ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Renk uzayını BGR'den RGB'ye dönüştürün
# Matplotlib figürüne mouse callback fonksiyonunu bağlayın
fig.canvas.mpl_connect('button_press_event', lambda event: mouse_callback(event, *event.xdata, *event.ydata, 0, None))
# Mouse callback fonksiyonunu Matplotlib figürüne bağlayın
fig.canvas.mpl_connect('button_press_event', mouse_callback)

# Pencereyi gösterin
plt.show()


```
